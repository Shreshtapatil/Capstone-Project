```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_movie_details(movie_title):
    search_url = f"https://www.themoviedb.org/search?query={movie_title}"
    response = requests.get(search_url)
    
    if response.status_code != 200:
        return {"title": movie_title, "error": "Failed to fetch data"}
    
    soup = BeautifulSoup(response.content, "html.parser")
    result = soup.find("div", class_="card v4 tight")
    if not result:
        return {"title": movie_title, "error": "No results found"}
    
    # Get movie details
    movie_link = f"https://www.themoviedb.org{result.find('a')['href']}" if result.find("a") else None
    movie_page = requests.get(movie_link)
    movie_soup = BeautifulSoup(movie_page.content, "html.parser")
    
    movie_details = {
        "movie_id": movie_link.split('/')[-1] if movie_link else None,
        "title_x": movie_title, 
        "cast": caste,  
        "crew": crew, 
        "budget": budget,  
        "genres": genre,  
        "homepage": homepagelink, 
        "id": movie_link.split('/')[-1] if movie_link else None,
        "keywords": keywords, 
        "original_language": laguage,  # O
        "overview": movie_soup.find('div', class_="overview").text.strip() if movie_soup.find('div', class_="overview") else "No overview available",
        "popularity": popularity,  
        "production_companies": Productioncompany, 
        "production_countries": country
        "release_date": date,
        "revenue": revenue,  
        "runtime": runtime,  
        "spoken_languages": dubbedin,  
        "status": status, 
        "tagline": tagline,  
        "title_y": movie_title 
        "vote_average": rating, 
        "vote_count": rating_count
    }

    return movie_details

def scrape_movies(num_movies=5000):
    movie_titles = []
    for i in range(1, num_movies + 1): 
        movie_titles.append(f"Movie {i}") 
    scraped_data = []
    
    for i, title in enumerate(movie_titles, start=1):
        if i > num_movies:
            break
        
        print(f"Scraping details for movie {i}/{num_movies}: {title}")
        details = scrape_movie_details(title)
        scraped_data.append(details)
        
        
        time.sleep(1)  
    
    return scraped_data

scraped_data = scrape_movies(5000)

output_path = r"C:\Users\Lenovo\Desktop\Python Project\scraped_movie_details.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)
df = pd.DataFrame(scraped_data)
df.to_csv(output_path, index=False)

print(f"Scraped data saved to: {output_path}")
```
